```python
# WILL CREATE A GUI TO GENERATE IMAGES USING CHATGPT API

import tkinter as tk
from PIL import Image, ImageTk
from openai import OpenAI
import requests
from io import BytesIO

client = OpenAI(api_key="Set your OpenAI API key")

def generate_image(prompt, model="dall-e-2", size="1024x1024", n=1, quality="standard"):
    try:
        # Call the OpenAI API to generate an image
        response = client.images.generate(
            model=model,
            prompt=prompt,
            size=size,
            quality=quality,
            n=n
        )

        # Get the image URL from the response
        image_url = response.data[0].url
        return image_url
    except Exception as e:
        print(f"Failed to generate image: {e}")
        return None

def display_image(image_url):
    try:
        # Load the image from the URL
        response = requests.get(image_url)
        img_data = response.content
        img = Image.open(BytesIO(img_data))
        img = img.resize((512, 512), Image.ANTIALIAS)
        img_tk = ImageTk.PhotoImage(img)

        # Display the image in the label
        img_label.config(image=img_tk)
        img_label.image = img_tk

        # Resize the window to fit the image
        root.geometry(f"512x650")  # Resize window to fit the image and button
    except Exception as e:
        print(f"Failed to display image: {e}")

def on_send():
    prompt = prompt_box.get("1.0", tk.END).strip()
    if prompt:
        model = "dall-e-3"  # Model choice, can be modified
        size = "1024x1024"  # Image size, can be modified
        quality = "standard"  # Image quality
        n = 1  # Number of images to generate
        
        image_url = generate_image(prompt, model=model, size=size, n=n, quality=quality)
        
        if image_url:
            display_image(image_url)
    else:
        print("Input Required: Please enter a prompt.")

# Create the GUI window
root = tk.Tk()
root.title("Architectural Co-Pilot")
root.geometry("512x150")

# Create a textbox for the prompt
prompt_box = tk.Text(root, height=4, width=50)
prompt_box.pack(pady=10)

# Create a send button
send_button = tk.Button(root, text="Send", command=on_send)
send_button.pack(pady=10)

# Label for displaying the generated image
img_label = tk.Label(root)
img_label.pack(pady=10)

# Start the GUI event loop
root.mainloop()
```
